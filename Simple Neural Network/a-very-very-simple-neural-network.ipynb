```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [(1,3), (3,4), (5,3.5)]
x = [d[0] for d in data]
y = [d[1] for d in data]
```

```python
# This is the line of best fit of the data
w0, b0 = np.polyfit(x,y,1)
y0 = [w0*xi+b0 for xi in x]
plt.scatter(x, y)
plt.plot(x, y0)
```

```python
costdata = []
def step(w=0, b=0, dw=0, db=0):
    # Forward Propagation
    a = []
    for xi in x:
        ai = w*xi + b
        a.append(ai)

    # Plot graph
    plt.plot(x, y0)
    plt.plot(x, a)
    plt.scatter(x,y)

    # Compute Cost
    cost = 0
    for i in range(len(a)):
        loss = (y[i]-a[i])**2
        cost += loss
    costdata.append(cost)

    # Backward Propagation
    innerw = []
    for i in range(len(x)):
        innerw.append( y[i] - w*x[i] - b )
        dw += -2*x[i]*innerw[i]


    innerb = []
    for i in range(len(x)):
        innerb.append( y[i] - w*x[i] - b )
        db += -2*innerb[i]

    # Update paramters
    w = w - 0.001*dw
    b = b - 0.001*db

    return w, b, dw, db
```

```python
# Initialize values
w, b, dw, db = step()
```

```python
# Rerun this cell multiple times to update the parameters,
# then run the cell bellow to visualize the cost
w, b, dw, db = step(w, b, dw, db)
```

```python
# Then run this cell to see the cost, then run the above cell
plt.plot(costdata)
```
