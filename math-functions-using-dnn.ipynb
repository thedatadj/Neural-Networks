In this notebook, I use neural network to model mathematical functions.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

# Linear Function
$y = 2x + 5$

```python
# Dataset
x = np.linspace(0,10,11)
y = 2*x + 5
```

```python
# Neural network

model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, input_shape=[1]),
    tf.keras.layers.Dense(1, activation='sigmoid')
    ])
```

```python
model.compile(optimizer='sgd',
              loss='mse')
```

```python
# Training
model.fit(x, y, epochs=10)
```

```python
y_pred = model.predict(x)
```

```python
plt.plot(x, y_pred, label='predicted')
plt.plot(x, y, label='true')
plt.legend()
```

```python
model.predict([25])
```

# Quadratic function

```python
# Dataset
y2 = x**2
```

```python
# Quadratic function graph
plt.plot(y2)
```

```python
# Let's see how well does the linear model do
linear_model = model
y2_pred = linear_model.predict(x)
```

```python
plt.plot(x, y2_pred)
plt.plot(x, y2)
```

```python
linear_model.evaluate(x, y2)
```

> The loss is so high.

```python
# Now let's train a network.
quadratic_model = tf.keras.Sequential([
    tf.keras.layers.Dense(1028, input_shape=[1]),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.Dense(512),
    tf.keras.layers.Dense(128),
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(32),
    tf.keras.layers.Dense(8),
    tf.keras.layers.Dense(4),
    tf.keras.layers.Dense(2),
    tf.keras.layers.Dense(1)
])
```

```python
quadratic_model.compile(optimizer='adam',
                        loss='mse')
```

```python
quadratic_model.fit(x, y2, epochs=300)
```

```python
y2_pred = quadratic_model.predict(x)
```

```python
plt.plot(x, y2_pred, label='predicted')
plt.plot(x, y2, label='true')
plt.legend()
```

```python
linear_cubic_pred = linear_model.predict(x)
quadratic_cubic_pred = quadratic_model.predict(x)
cubic_cubic_pred = cubic_model.predict(x)
```

# Sine function

```python
# Dataset
x = np.linspace(0, 10, 301)
y = np.sin(x)
plt.plot(x, y)
```

```python
# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2056, input_shape=[1]),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.Dense(1028),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(512),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(128),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(64),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(8),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(4),
    tf.keras.layers.Dense(2),
    tf.keras.layers.Dense(1)
])

class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if (logs.get('loss') < 0.01):
            print("\nReached specified loss so cancelling training!")
            self.model.stop_training = True

callbacks = myCallback()

model.compile(
    optimizer='adam',
    loss='mse'
)


model.fit(x, y, epochs=200, callbacks=[callbacks])
```

```python
y_sin = model.predict(x)
```

```python
plt.plot(x, y_sin, label='prediction')
plt.plot(x, y, label='true')
plt.legend()
```
