In this notebook I create a model to predict the compressive strength of concrete manufactured according to various recipes.

```python
import pandas as pd

# Load data
data = pd.read_csv("/content/concrete.csv")
data.head()
```

> The target variable is `CompressiveStrength`.

# Data preparation
I prepare the data for training.

```python
X = data.copy()
y = X.pop('CompressiveStrength')
```

```python
# Data split
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=0)
```

# Modeling
I create a model for the prediction task.

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Model arquitecture
model = Sequential([
    Dense(16, activation='relu', input_shape=[8]),
    Dense(128, activation='relu'),
    Dense(512, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1)
])

# Compiling
model.compile(optimizer='adam',
              loss='mse')

# Model training
model.fit(X_train, y_train, epochs=100)
```

# Evaluation

```python
model.evaluate(X_valid, y_valid)
```
